```python
import pandas as pd
import re
from nltk.corpus import stopwords

def clean_text(text):
    # Remove URLs
    text = re.sub(r'http\S+', '', text)

    # Remove digits
    text = re.sub(r'\d+', '', text)

    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]

    return ' '.join(filtered_words)

# Load the CSV file into a DataFrame
input_file = 'sentiment_analysis.csv'
output_file = 'cleaned_dataset.csv'
df = pd.read_csv(input_file)

# Clean the 'text' column using the clean_text function
df['cleaned_text'] = df['text'].apply(clean_text)

# Save the cleaned DataFrame to a new CSV file
df.to_csv(output_file, index=False)

print("Dataset cleaned and saved to:", output_file)
```

```python
import pandas as pd

# Load the CSV file into a DataFrame
input_file = 'sentimentdataset.csv'
output_file = 'transformed_dataset.csv'
df = pd.read_csv(input_file)

# Remove hash symbol from hashtags in the 'text' column
df['transformed_text'] = df['Hashtags'].apply(lambda x: x.replace('#', '') if '#' in x else x)

# Save the transformed DataFrame to a new CSV file
df.to_csv(output_file, index=False)

print("Dataset transformed and saved to:", output_file)
```
